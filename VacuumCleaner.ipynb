1. Agent program for two states vacuum environment.

```python
import random

#setting up environment
def environment():
  room_condition = {'A':0,'B':0}
  room_condition['A'] = random.randint(0,1)
  room_condition['B'] = random.randint(0,1)

  print("Initial Environment condition:")

  for room, condition in enumerate(room_condition):
    print(f"Room {room}: {condition}")

  return room_condition

def agent(envi):
  rooms = ['A','B']
  loc = random.choice(rooms)
  other_loc = rooms[rooms.index(loc) - 1]

  print("Vacuum cleaner agent is randomly placed in room {}".format(loc))

  print("     Condition       |     Action      ")
  print("---------------------------------------")

  if envi[loc] == 0 :
    print('Room {} is dirty.      Cleaning room {}'.format(loc,loc))
    envi[loc] = 1
    print('Room {} is clean.      Moving to room {}'.format(loc,other_loc))

    if envi[other_loc] == 0:
      print('Room {} is dirty.      Cleaning room {}'.format(other_loc,other_loc))
      envi[loc] = 1
      print('Both rooms are now cleaned    :)!')
    elif envi[other_loc] == 1:
      print('Room {} is clean.      Both rooms are now cleaned :)!'.format(other_loc))

  elif envi[loc] == 1:
    print('Room {} is clean.      Moving to room {}'.format(loc,other_loc))

    if envi[other_loc] == 0:
      print('Room {} is dirty.      Cleaning room {}'.format(other_loc,other_loc))
      envi[loc] = 1
      print('Both rooms are now cleaned    :)!')
    elif envi[other_loc] == 1:
      print('Room {} is clean.      Both rooms are now cleaned :)!'.format(other_loc))


envi = environment()
agent(envi)
```

2. Table driven agent

```python
import random

def table_driven_agent(percept, table):
    action = table.get(percept)
    return action

def main():
    # Define the table for the Table-Driven Agent
    table = {
        ('A', 'dirty'): 'suck',
        ('A', 'clean'): 'move_right',
        ('B', 'dirty'): 'suck',
        ('B', 'clean'): 'move_left'
    }

    rooms = ['A', 'B']
    environment_condition = {room: 'clean' if random.random() < 0.5 else 'dirty' for room in rooms}

    print("Initial Environment Condition:")
    for room in rooms:
        print(f"Room {room}: {environment_condition[room]}")

    agent_location = random.choice(rooms)
    other_location = rooms[rooms.index(agent_location) - 1]

    print("\nVacuum cleaner agent is randomly placed in room", agent_location,"\n")
    print("     Condition       |     Action      ")
    print("---------------------------------------")

    while 'dirty' in environment_condition.values():
        percept = (agent_location, environment_condition[agent_location])
        action = table_driven_agent(percept, table)

        if action == 'suck':
            print(f"Room {agent_location} is dirty          Suck dirt")
            environment_condition[agent_location] = 'clean'

        elif action == 'move_right':
            print(f"Room {agent_location} is clean          Move Right")
            agent_location = other_location

        elif action == 'move_left':
            print(f"Room {agent_location} is clean          Move Left")
            agent_location = other_location

    print("\nBoth rooms are now cleaned :)!")


if __name__ == "__main__":
    main()
```

4. Vacuum cleaner working 8 hour shift in n*m environment

```python
import random

n = random.randint(8, 15)
m = random.randint(8, 15)

def environment():
    room = []
    for i in range(n):
        row = [1 if random.random() < 0.06 else 0 for _ in range(m)]
        room.append(row)

    print(f'n*m ({n}*{m}) environment is:\n')
    # Printing the environment
    for i in room:
        for j in i:
            print("|{}|".format(j), end="")
        print("\n", end="")
        print("-" * 3 * m)

    print("1 represents a clean square.\n0 represents a dirty square.\n")
    return room

def action(loc, room, score):
    room[loc[0]][loc[1]] = 0
    score = score + 1
    print("Room [", loc[0], ",", loc[1], "] is cleaned")
    return score

def move(n, m, room, score):
    print("Vacuum cleaner agent is ready to clean the room!!\n")
    for i in range(n):
        for j in range(m):
            if room[i][j] == 1:
                score = action([i, j], room, score)
    return score

def main():
    score = 0  # Initialize the score variable
    print("Hardworking vacuum cleaner agent's hour shift:\n\n")
    for i in range(1, 9):
        room = environment()
        score = move(n, m, room, score)
        print("\n\n------{} HOUR DONE---------\n\n".format(i))

    print("PERFORMANCE MEASURE:")
    # Performance measure is percent of total number of clean tiles
    pm = (((n * m * 8) - score) / (n * m * 8)) * 100
    print(f"score = {pm:.2f}%")

if __name__ == "__main__":
    main()
```
